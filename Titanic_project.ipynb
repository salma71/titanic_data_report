```python
def male_female_child(passenger):
    age,sex = passenger
    
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
titanic_df['person'] = titanic_df[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
#kd plot
#draw a multiple grids using facetgrid
fig = sns.FacetGrid(titanic_df,hue='Sex',aspect=4)

fig.map(sns.kdeplot,'Age',shade=True)

#set the oldest age to the max
oldest = titanic_df['Age'].max()

#since the age cannot be negative, so I set the limits from 0 to the oldest
fig.set(xlim=(0,oldest))

#finally I add a legend
fig.add_legend()
```

```python
# we need to remove the NaN from the cabin column
deck = titanic_df['Cabin'].dropna()
deck.head()
```

```python
# to grap the letters only from the cabin name
#create an empty array
levels = []

#loop over the deck list(the list without NaN value) and append the first letter only to that empty array.
for level in deck:
    levels.append(level[0])
    
cabin_df = DataFrame(levels)
cabin_df.columns = ['Cabin']

sns.factorplot('Cabin',kind='count',data=cabin_df,palette='autumn_d', order=['A','B','C','D','E','F','G'])
```

```python
cabin_df = cabin_df[cabin_df.Cabin != 'T']#to remove T 
sns.factorplot('Cabin',kind='count',data=cabin_df,palette='summer', order=['A','B','C','D','E','F','G'])
```

```python
#now we have to determine who was alone and who was with the family
#define a new column called alone 
titanic_df['alone'] = titanic_df.SibSp + titanic_df.Parch
#this means it the column is zero so the passenger is compeletly alone on board.

titanic_df['alone'].head()
```

```python
titanic_df['alone'].loc[titanic_df['alone'] > 0] = 'With family'

titanic_df['alone'].loc[titanic_df['alone'] == 0] = 'The passenger is alone'
```
